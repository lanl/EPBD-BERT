```python
import re
import gzip
import pyfastx
home_dir="../"
```

```python
human_genome_filepath = home_dir+"data/downloads/hg19_latest/hg19.fa.gz"
fa = pyfastx.Fasta(human_genome_filepath)
fa
```

```python
search = re.compile(r'[^ACGT]').search
def any_char_other_than_ACGT_in_seq(seq):
    """returns True if any other char found in the input seq than {A,C,G,T} (uppercase only).
    """
    return bool(search(seq))
```

```python
# this block takes more than 2hours to run.
f_out = open(home_dir+"data/processed/peaks_with_labels_issues.tsv", "wt")

with gzip.open(home_dir+"data/processed/peaks_with_labels.tsv.gz", 'rt') as f:
    i=0
    j=0 # the number of 200 bp bins that contains chars other than ACGT.
    for line in f:

        # print(line.split(sep="\t"))
        chrom, start, end, n_labels, labels = line.split(sep="\t")
        start = int(start.strip())
        end = int(end.strip())

        seq = fa[chrom][start:end].seq
        seq = seq.upper()
        # print(i, seq)

        if any_char_other_than_ACGT_in_seq(seq):
            j+=1
            f_out.write(line)


        i+=1
        # if i==10: break
        if i%10000==0:
            print(i, j)

f_out.close()

i, j
```
