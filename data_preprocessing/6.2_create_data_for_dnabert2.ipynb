```python
import gzip
import pyfastx
home_dir="../"
```

```python
human_genome_filepath = home_dir+"data/downloads/hg19_latest/hg19.fa.gz"
fa = pyfastx.Fasta(human_genome_filepath)
fa
```

```python
flank_size=400
seq_dict = {} # {key: chr_start_end, value: seq}
with gzip.open(home_dir+"data/processed/peaks_with_labels_clean.tsv.gz", 'rt') as f:
    i=0
    for line in f:
        chrom, start, end, n_labels, labels = line.split(sep="\t")
        start = int(start.strip())
        end = int(end.strip())

        seq = fa[chrom][start-flank_size:end+flank_size].seq
        seq = seq.upper()
        # print(len(seq))
        seq_dict[f"{chrom}_{str(start)}_{str(end)}"] = seq
        
        i+=1
        # if i==10: break
        if i%10000==0: 
            print(i)
```

```python
len(seq_dict) # flank_seq+seq+flank_seq = 400+200+400
```

```python
import pickle
with open(home_dir+"data/processed/seq_with_flanks_dict.pkl", "wb") as f:
    pickle.dump(seq_dict, f)
```
