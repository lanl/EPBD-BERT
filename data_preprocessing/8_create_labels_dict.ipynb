```python
import os
import sys
home_dir = "../"
module_path = os.path.abspath(os.path.join(home_dir))
if module_path not in sys.path:
    sys.path.append(module_path)

import pandas as pd
import utils as pickle_utils
```

```python
# reading and processing peakfiles metadata to generate the labels dict
col_names = ["filename", "project", "lab", "composite", "dataType", "view", "cell", "treatment", "antibody", "control", "dataVersion", "dccAccession", "controlId", "quality", "tableName", "type", "md5sum", "size"]
with open(home_dir+"data/downloads/wgEncodeAwgTfbsUniform/files.txt", "r") as h:
    data = []
    for line in h.readlines():
        # print(line.strip().split(";"))
        line_items = line.strip().split(";")

        row = {}
        filename, project = line_items[0].split("\t")
        project = project.split("=")[1]
        # print(filename, project)

        row["filename"] = filename
        row["project"] = project

        for line_item in line_items[1:]:
            key, value = line_item.strip().split("=")
            row[key] = value

        data.append(row)
        # print(row)
        # break
processed_peaks_metadata_df = pd.DataFrame.from_records(data)
processed_peaks_metadata_df.head(3)
```

```python
# creating peakfiles vs index (0-based) dict for generating labels
peak_file_names = processed_peaks_metadata_df["tableName"].unique()
peak_file_names.sort()
peakfilename_index_dict = {x:i for i, x in enumerate(peak_file_names)} # 0-indexed
print(len(peakfilename_index_dict))
```

```python
pickle_utils.save_as_pickle(peakfilename_index_dict, home_dir+"data/processed/peakfilename_index_dict.pkl")
```
